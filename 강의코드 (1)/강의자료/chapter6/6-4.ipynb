## (2) 일별 주가 데이터 가져오기

- get_market_ohlcv: 전체 시장의 주가 데이터를 한번에 가져오는 방법. 한 번에 한 날짜씩 가져올 수 있음

```python
# 오늘 날짜 가져오기

from datetime import datetime
date = str(datetime.now())[:10]
```

```python
# 오늘 일별 주가 데이터 가져오기

stock.get_market_ohlcv(date,market='KOSPI')
```

## (3) 시가총액 데이터

- 해당 날짜의 종목별 시가총액을 내림차순으로 리턴

```python
# 오늘 시가총액 데이터 가져오기

stock.get_market_cap(date)
```
