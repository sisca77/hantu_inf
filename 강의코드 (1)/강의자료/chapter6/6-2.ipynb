2. 접근토큰 발급 기능 구현(https://apiportal.koreainvestment.com/apiservice/oauth2#L_fa778c98-f68d-451e-8fff-b1c6bfe5cd30)

```python
# api key 불러오기

import yaml

with open('config.yaml', 'r') as f:
    config = yaml.load(f, Loader=yaml.FullLoader)

api_key = config['hantu']['api_key']
secret_key = config['hantu']['secret_key']
```

```python
import requests
import json

base_url = 'https://openapi.koreainvestment.com:9443'

headers = {"content-type":"application/json"}
body = {
        "grant_type":"client_credentials",
        "appkey":api_key, 
        "appsecret":secret_key,
        }
url = base_url + '/oauth2/tokenP'
res = requests.post(url, headers=headers, data=json.dumps(body)).json()
```

#### 접근토큰 설명

- access_token: 발급받은 접근토큰. 이 값을 저장해두고 사용하면 됨
- access_token_token_expired: 접근토큰 만기시점. 발급 순간부터 24시간 후
- token_type: 접근토큰유형. 'Bearer' 값이 들어옴
- expires_in: 만기까지 남은 시간(초)

# 2. 주가 데이터 가져오기

```python
access_token = res['access_token']
```

```python
ticker = '005930'

tr_id = 'FHKST01010100'

headers = {"content-type":"application/json",
                "appkey":api_key, 
                "appsecret":secret_key,
                "authorization":f"Bearer {access_token}",
                "tr_id":tr_id,
                }

params = {
        "fid_cond_mrkt_div_code":"J",
        "fid_input_iscd": ticker
        }
url = base_url + '/uapi/domestic-stock/v1/quotations/inquire-price'

response = requests.get(url, headers=headers, params=params)
```
